## Project Overview
This project demonstrates exploratory data analysis (EDA) and supervised classification on the Iris dataset using two machine learning algorithms:

- Logistic Regression
- K-Nearest Neighbors (KNN)

The objective is to compare both models and evaluate their performance on a structured and balanced dataset.

```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Model building
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

#### Load the dataset

```python
data = pd.read_csv("Iris_data_sample.csv")
data.head()
```

#### Checking Missing Values

```python
data.isnull().sum()
```

The dataset is checked for missing values across all columns.

A small number of missing values were identified and handled using mean imputation for numerical features and mode imputation for categorical features.

```python
data['SepalLengthCm'] = pd.to_numeric(data['SepalLengthCm'], errors='coerce')
data['PetalLengthCm'] = pd.to_numeric(data['PetalLengthCm'], errors='coerce')
```

#### Data Type Conversion

Some columns are converted to numeric format to ensure compatibility with statistical analysis and machine learning algorithms.  
Invalid entries (if any) were coerced into NaN values and handled appropriately.  

```python
# Removing Irrelevant Columns
data = data.drop(columns=['Unnamed: 0'])

# The column 'Unnamed: 0' is removed as it represents an index column and does not contribute to model performance or analysis.
```

```python
data['SepalLengthCm'] = data['SepalLengthCm'].fillna(data['SepalLengthCm'].mean())
data['SepalWidthCm'] = data['SepalWidthCm'].fillna(data['SepalWidthCm'].mean())
data['PetalLengthCm'] = data['PetalLengthCm'].fillna(data['PetalLengthCm'].median())
```

#### Handling Missing Values

- Missing numerical values were identified in the dataset.
- For most features, mean imputation was considered appropriate since their distributions appear relatively symmetric.
- However, for **PetalLengthCm**, a noticeable difference between the mean and median was observed.  
  This suggests the presence of outliers or skewness in the data.

Therefore, median imputation was used for *PetalLengthCm* as it is more robust to extreme values and provides a more reliable central tendency measure in skewed distributions.

```python
data['Species'].value_counts()
```

```python
data['Species'] = data['Species'].fillna(data['Species'].mode()[0])
```

#### Handling Missing Values (Categorical Feature)

One missing value was observed in the 'Species' column.

Since 'Species' is a categorical variable, mode imputation was used to replace the missing value with the most frequently occurring class.

```python
data.isnull().sum()
```

- After completing numerical and categorical imputations, the dataset is rechecked.
- All missing values have now been successfully handled, and the dataset is ready for exploratory data analysis.

### Model Preparation

Before training the models, the dataset is divided into:

- **Features (X):** Independent variables used for prediction  
- **Target (y):** Dependent variable representing the species  

This separation is required for supervised machine learning.

```python
X = data.drop("Species", axis=1)
y = data["Species"]
```

```python
X.head()
```

### Trainâ€“Test Split

The dataset is split into training and testing sets to evaluate model performance on unseen data.

A stratified split is used to maintain balanced class distribution across both sets.

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
y_train.value_counts()
```

1. The training set maintains a balanced distribution across all three species.   
2. Stratified splitting ensures that each class is proportionally represented in both training and testing sets, preventing bias during model evaluation.

### Feature Scaling

Feature scaling is applied to standardize the numerical features. 

This is especially important for distance-based algorithms like K-Nearest Neighbors, where differences in feature scale can disproportionately influence distance calculations.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Logistic Regression

Logistic Regression is a supervised learning algorithm used for multi-class classification.

It models the probability of each class using a linear decision boundary.

This model serves as a strong baseline for evaluating classification performance on structured and well-separated datasets such as Iris.

```python
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train,y_train)

y_pred = log_reg.predict(X_test)
```

```python
print("Logistic Regression Performance Metrics")
print("----------------------------------------")

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

### Logistic Regression Interpretation

The Logistic Regression model achieved an accuracy of approximately **94.7%** on the test set.

This indicates that the model is able to correctly classify the majority of unseen samples, demonstrating strong predictive performance on the Iris dataset.

---

The evaluation metrics indicate strong overall performance.

- The confusion matrix shows that most samples are correctly classified.
- The classification report provides class-wise precision, recall, and F1-score, confirming balanced performance across species.

Overall, Logistic Regression performs effectively on this dataset.

### K-Nearest Neighbors (KNN)

K-Nearest Neighbors is a distance-based classification algorithm.

It classifies a data point based on the majority class among its *k* nearest neighbors in the feature space.

Since KNN relies on distance calculations, feature scaling applied earlier is particularly important for this model.

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

y_pred_knn = knn_model.predict(X_test)
```

```python
print("KNN Performance Metrics")
print("-----------------------")

print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("\nClassification Report:\n", classification_report(y_test, y_pred_knn))
```

### KNN Interpretation

The KNN model achieved an accuracy of approximately **92.1%** on the test set.

The confusion matrix indicates that most samples were correctly classified, with minor misclassification occurring between Versicolor and Virginica. This is expected, as these two species exhibit overlapping feature characteristics.

---

```python
comparison = pd.DataFrame({
    "Model": ["Logistic Regression", "K-Nearest Neighbors"],
    "Accuracy": [accuracy_score(y_test, y_pred), accuracy_score(y_test, y_pred_knn)]
})
comparison.sort_values(by="Accuracy", ascending=False)
comparison
```

## Final Model Comparison

Logistic Regression achieved slightly higher accuracy compared to KNN on this dataset.

Both models performed strongly due to the structured and well-separated nature of the Iris dataset. However, Logistic Regression demonstrated marginally better generalization on unseen data.

This comparison highlights how both linear and distance-based algorithms can be effective when features exhibit clear separability.
